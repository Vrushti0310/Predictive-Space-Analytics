```python
#4 Data Cleaning

df.isnull().sum()
df.isnull().values.any()    # Quick True/False check
```

```python
# Option 1 - Fill with median (recommended)
df['Mission Cost (billion USD)'] = df['Mission Cost (billion USD)'].fillna(df['Mission Cost (billion USD)'].median())

# Option 2 - Use np.where to flag missing costs
df['cost_missing'] = np.where(df['Mission Cost (billion USD)'].isnull(), 1, 0)
```

```python
print(df.duplicated().sum())
df = df.drop_duplicates()         # Remove duplicate rows
```

```python
# Fix the Date Column
df['Launch Date'] = pd.to_datetime(df['Launch Date'], utc=True)
df['Year'] = df['Launch Date'].dt.year
df['Decade'] = (df['Year'] // 10) * 10
```

```python
#Clean MissionStatus — Target Variable

print(df['MissionStatus'].value_counts())

# Simplify to binary: Success = 1, Failure = 0
df['MissionSuccess'] = np.where(
    df['MissionStatus'].str.contains('Success', case=False), 1, 0
)
```
